```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(data= cancer['data'], columns=cancer['feature_names'])
df['positive'] = cancer['target']
```

```python
df
```

```python
df['positive'].value_counts()
```

```python
df_pos= df[df['positive']==1]
```

```python
df_neg= df[df['positive']==0]
```

```python
df_neg
```

```python
df_neg_20 = df_neg[:20]
```

```python
cancer = pd.concat([df_neg_20,df_pos])
```

```python
cancer['positive'].value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(cancer.iloc[:,:-1],cancer.iloc[:,-1],test_size=0.2)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

```python
model = LogisticRegression()
```

```python
standard_scaler = preprocessing.StandardScaler()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
one_pred = np.ones(shape=y_test.shape)
```

```python
metrics.accuracy_score(y_test,one_pred)
```

```python
metrics.precision_score(y_test,pred)
```

```python
metrics.precision_score(y_test,one_pred)
```

```python
metrics.recall_score(y_test,pred)
```

```python
metrics.recall_score(y_test,one_pred)
```

```python
metrics.f1_score(y_test,pred)
```
